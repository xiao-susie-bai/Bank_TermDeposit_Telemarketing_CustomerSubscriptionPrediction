```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = pd.read_csv("/Users/baixiao/Desktop/X_balance.csv")
X.head()
```

```python
X['age_log'] = np.log(X['age'])
X['campaign_log'] = np.log(X['campaign'])
#X
```

```python
X['pdays_log'] = np.log(X['pdays']+2)
X['previous_log'] = np.log(X['previous']+2)
#X.head()
```

```python
X['cons.price.idx_trans'] = np.sqrt(X['cons.price.idx']+1-X['cons.price.idx'].min())
X.head()
```

```python
X['job_encoded_log'] = np.log(X['job_encoded']+2)
X['housing_encoded_log'] = np.log(X['housing_encoded']+2)
X['loan_encoded_log'] = np.log(X['loan_encoded']+2)
X['contact_encoded_log'] = np.log(X['contact_encoded']+2)
X.head()
```

plotting to check the distribution after these transformation

```python
import numpy as np
import pylab
import scipy.stats as stats
```

```python
for i in ['age_log', 'campaign_log', 'pdays_log', 'previous_log', 'cons.price.idx_trans', 'job_encoded_log', 'housing_encoded_log', 'loan_encoded_log', 'contact_encoded_log']:
    stats.probplot(X[i], dist="norm", plot=pylab)
    pylab.title("QQ-plot of "+str(i))
    pylab.show()
```

```python
for i in ['age_log', 'campaign_log', 'pdays_log', 'previous_log', 'cons.price.idx_trans', 'job_encoded_log', 'housing_encoded_log', 'loan_encoded_log', 'contact_encoded_log']:
    plt.hist(X[i])
    plt.title("QQ-plot of "+str(i))
    plt.show()
```
